```python
import signalz
import matplotlib.pyplot as plt
import numpy as np

gen_len = 10000
trans_len = 100
mg = signalz.mackey_glass(gen_len+trans_len, a=0.2, b=1, c=0.9, d=17, e=10, initial=0.1)[trans_len:]
mg = (mg - min(mg))/(max(mg)-min(mg))

np.savetxt('./mg17.csv', mg)

fig = plt.figure(figsize=(10, 3))
plt.plot(mg[:gen_len//10])
plt.show()
```
